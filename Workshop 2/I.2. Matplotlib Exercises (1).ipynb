## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
# Create figure object
fig = plt.figure()

# Add axes to the figure at position [x0, y0, width, height]
ax = fig.add_axes([0, 0, 1, 1])

# Plot x vs y
ax.plot(x, y)

# Set labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Title')

# Display the plot
plt.show()
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Create figure object
fig = plt.figure()

# Add first axes (big plot)
ax1 = fig.add_axes([0, 0, 1, 1])

# Add second axes (small inset plot)
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

# Plot x vs y on both axes
ax1.plot(x, y)
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.set_title('Larger Plot')

ax2.plot(x, y)
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
ax2.set_title('Smaller Plot')

# Display the figure
plt.show()
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Create the figure and main axes
fig = plt.figure(figsize=(6, 4))
main_ax = fig.add_axes([0, 0, 1, 1])

# Main plot: x vs z
main_ax.plot(x, z, 'b')  # blue line
main_ax.set_xlabel("X", fontweight='bold')
main_ax.set_ylabel("Z", fontweight='bold')

# Inset plot: zoomed-in x vs y
inset_ax = fig.add_axes([0.2, 0.5, 0.35, 0.35])  # [left, bottom, width, height]
inset_ax.plot(x, y, 'b')
inset_ax.set_xlim(20, 22)
inset_ax.set_ylim(35, 45)
inset_ax.set_title("zoom")
inset_ax.set_xlabel("X")
inset_ax.set_ylabel("Y")

plt.show()
```

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
plt.show()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))
ax1.plot(x, y, 'r--')  # red line
ax1.set_title("x vs y")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

# Second subplot: x vs z
ax2.plot(x, z, 'b')  # blue line
ax2.set_title("x vs z")
ax2.set_xlabel("x")
ax2.set_ylabel("z")
plt.show()
```
